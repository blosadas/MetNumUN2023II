Polinomio en Sympy:

```
from sympy import Symbol
from sympy import div

x = Symbol('x')

p = x**10 - 15*x**9 - 8*x**8 + 1094*x**7 - 6239*x**6 + 9821*x**5 + 18318*x**4 - 70804*x**3 + 39768*x**2 + 59904*x - 51840

roots = [(-1), 10, 3, (-2), 4, (-9), 4, 2, 3, 1]

for i in roots:
  p, r = div(p, x-i)
  print(f"Raiz: (x-{i}): ")
  print(f"Cociente: {p}")
  print(f"Residuo: {r}\n")
```

## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)


Sympy Symbolic

```
from sympy import symbols, I, N, pi, expand
y1, y2, z1, z2, x1, x2 = symbols("x1 x2 y1 y2 z1 z2", real=True)
x = 69*x1 + I*x2
y = 420*y1 + I*y2
z = 13*z1 + I*z2

print(x*y)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
b=1/(2*pi + I)
print('b=',b)
b=N(b)
print('b=',b)
```
