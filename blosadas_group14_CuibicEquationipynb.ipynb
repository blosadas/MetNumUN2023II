```python
def cubic(b, c, d, x):
    v = x**3 + b*x**2 + c*x + d
    return v
```

```python
import math
import numpy as np

def solve_cubic(b, c, d):
    a = 1
    if (a == 0 and b == 0):
        return np.array([(-d * 1.0) / c])

    elif (a == 0):
        D = c * c - 4.0 * b * d
        if D >= 0:
            D = math.sqrt(D)
            x1 = round((-c + D) / (2.0 * b))
            x2 = round((-c - D) / (2.0 * b))
        else:
            D = math.sqrt(-D)
            x1 = round((-c + D * 1j) / (2.0 * b))
            x2 = round((-c - D * 1j) / (2.0 * b))

        return x1, x2

    f = findF(a, b, c)
    g = findG(a, b, c, d)
    h = findH(g, f)

    if f == 0 and g == 0 and h == 0:
        if (d / a) >= 0:
            x = round((d / (1.0 * a)) ** (1 / 3.0) * -1)
        else:
            x = round((-d / (1.0 * a)) ** (1 / 3.0))
        return x, x, x

    elif h <= 0:
        i = math.sqrt(((g ** 2.0) / 4.0) - h)
        j = i ** (1 / 3.0)
        k = math.acos(-(g / (2 * i)))
        L = j * -1
        M = math.cos(k / 3.0)
        N = math.sqrt(3) * math.sin(k / 3.0)
        P = (b / (3.0 * a)) * -1

        x1 = round(2 * j * math.cos(k / 3.0) - (b / (3.0 * a)))
        x2 = round(L * (M + N) + P)
        x3 = round(L * (M - N) + P)

        return x1, x2, x3

    elif h > 0:
        R = -(g / 2.0) + math.sqrt(h)
        if R >= 0:
            S = R ** (1 / 3.0)
        else:
            S = (-R) ** (1 / 3.0) * -1
        T = -(g / 2.0) - math.sqrt(h)
        if T >= 0:
            U = (T ** (1 / 3.0))
        else:
            U = ((-T) ** (1 / 3.0)) * -1

        x1 = round((S + U) - (b / (3.0 * a)))
        x2 = -(S + U) // 2 - (b // (3.0 * a)) + (S - U) * math.sqrt(3) * 0.5j
        x3 = -(S + U) // 2 - (b // (3.0 * a)) - (S - U) * math.sqrt(3) * 0.5j


        return x1, x2, x3

def findF(a, b, c):
    return ((3.0 * c / a) - ((b ** 2.0) / (a ** 2.0))) / 3.0

def findG(a, b, c, d):
    return (((2.0 * (b ** 3.0)) / (a ** 3.0)) - ((9.0 * b * c) / (a **2.0)) + (27.0 * d / a)) /27.0

def findH(g, f):
    return ((g ** 2.0) / 4.0 + (f ** 3.0) / 27.0)

```

```python
import cmath  # Importing complex math module for handling complex roots

def solve_cubic_equation( b, c, d):
    a = 1
    # Calculate the discriminant and other intermediate values
    delta_0 = b**2 - 3*a*c
    delta_1 = 2*b**3 - 9*a*b*c + 27*a**2*d
    C = ((delta_1 + cmath.sqrt(delta_1**2 - 4*delta_0**3)) / 2)**(1/3)

    # Calculate the roots using the cubic formula
    root_1 = -(1/(3*a)) * (b + C + delta_0 / C)
    root_2 = -(1/(3*a)) * (b + cmath.exp(2j * cmath.pi / 3) * C + delta_0 / (C * cmath.exp(2j * cmath.pi / 3)))
    root_3 = -(1/(3*a)) * (b + cmath.exp(4j * cmath.pi / 3) * C + delta_0 / (C * cmath.exp(4j * cmath.pi / 3)))

    return root_1, root_2, root_3

# Example usage:
b = -6
c = 11
d = -6

roots = solve_cubic_equation(b, c, d)
print("Roots:", roots)
```

```python
# Case 1: Three real different roots
b1, c1, d1 = -6, 11, -6
roots1 = solve_cubic(b1, c1, d1)

# Case 2: Roots at 0, 1, 2
b2, c2, d2 = 0, 1, 2
roots2 = solve_cubic(b2, c2, d2)

# Case 3: Root of multiplicity two
b3, c3, d3 = -6, 12, -8
roots3 = solve_cubic(b3, c3, d3)

# Case 4: Complex roots with small b
b4, c4, d4 = 1, 0, 1
roots4 = solve_cubic(b4, c4, d4)

# Case 5: Complex roots with large b
b5, c5, d5 = -1000, 2000, -1000
roots5 = solve_cubic(b5, c5, d5)

#Case 6: x**3
b6, c6, d6 = 0, 0, 0
roots6 = solve_cubic(b6, c6, d6)

#Case 7: x**3 -3*x**2 + 3*x -1
b7, c7, d7 = -3, 3, -1
roots7 = solve_cubic(b7, c7, d7)

#Case 8: x**3 - x**2 + x - 1
b8, c8, d8 = -1, 1, -1
roots8 = solve_cubic(b8, c8, d8)

# Print results
print("Case 1:", roots1)
print("Case 2:", roots2)
print("Case 3:", roots3)
print("Case 4:", roots4)
print("Case 5:", roots5)
print("Case 6:", roots6)
print("Case 7:", roots7)
print("Case 8:", roots8)
```

```python
import matplotlib.pyplot as plt
x = np.linspace(-2, 2, 400)

cases = [
    (-6, 11, -6),
    (0, 1, 2),
    (-6, 12, -8),
    (1, 0, 1),
    (-1000, 2000, -1000),
    (0, 0, 0),
    (-3, 3, -1),
    (-1, 1, -1)
]

plt.figure(figsize=(16, 12))
plt.rc('font', size=10)

for i, (b, c, d) in enumerate(cases):
    plt.subplot(3, 3, i + 1)
    roots = solve_cubic(b, c, d)
    y = cubic(b, c, d, x)
    plt.plot(x, y, label=f'Roots: {roots}')
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Case {i + 1}')
    plt.legend()

plt.tight_layout()
plt.show()
```
